```python
def get_movie_id (movies):
    
    movie_ids = {} # Make dict 
    
    for movie_name in movies: # For each movie in movies list
        # Created the variable for find id url.
        url = "https://imdb146.p.rapidapi.com/v1/find/"
        querystring = {"query":movie_name} 

        headers = {
        	"X-RapidAPI-Key": "eff4b077aamsh1b1feefcc13c29dp1a338ajsn161f7f62b8ea",
        	"X-RapidAPI-Host": "imdb146.p.rapidapi.com"
            }

        response = requests.get(url, headers=headers, params=querystring)
        
        
        if response.status_code == 200: # Is a condition that verify if the request was succesfull. 
            data = response.json()
            movie_id = data["titleResults"]["results"][0]["id"]
            movie_ids[movie_name] = movie_id # Make the dictionary. 
            import json
            
        else:
            print("None")

    with open("data.json", "w") as json_file: # To save the information. 
        json.dump(movie_ids, json_file)
        
    return movie_ids
```

```python
movie_ids = get_movie_id(movies)
movie_ids
```

```python
# Fuction to obtain more info from de id. 

def title_description(movie_ids):
    title_info = {}
    for key, movie_id in movie_ids.items():
        # The url for find title information. 
        url2 = "https://imdb146.p.rapidapi.com/v1/title/"
    
        querystring = {"id":movie_id}
    
        headers = {
            "X-RapidAPI-Key": "eff4b077aamsh1b1feefcc13c29dp1a338ajsn161f7f62b8ea",
        	"X-RapidAPI-Host": "imdb146.p.rapidapi.com"
            }

        response = requests.get(url2, headers=headers, params=querystring)

        if response.status_code == 200:
            data = response.json()
            try:
                director = data["directors"][0]["credits"][0]["name"]["nameText"]["text"] 
                genre = data["genres"]["genres"][0]["text"]
                title_info[movie_id] = {
                        "title": key,
                        "director": director,
                        "genres": genre
                    }
            except:
                title_info[movie_id] = {
                    "title": key,
                    "director": "not found",
                    "genres": "not found"
                }
                
        else:
            title_info[movie_id] = {
                    "title": key,
                    "director": "not found",
                    "genres": "not found"
                }
    
    return title_info
```

```python
movie_info = title_description(movie_ids)
movie_info
```

```python
# Make DataFrame
df2 = pd.DataFrame.from_dict(movie_info, orient='index')

# Print DataFrame
df2
```
