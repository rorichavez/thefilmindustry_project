```python
# To chance a specific element in a column
api.at[4, "genres"] = "Action"
api.at[4, "director"] = "Joaquim Dos Santos"
api.at[15, "genres"] = "Action"
api.at[31, "director"] = "David Fincher"
api.at[31, "genres"] = "Suspense"
```

# MERGE

```python
df = web.merge(api, left_index=True, right_index=True)
```

```python
# Delete one of the columns title. 
df = df.drop(['title_y'], axis=1)

# Change column name. 
df.rename(columns={'title_x': 'title'}, inplace=True)

# Start index with number 1. 
df.index = range(1, len(df) + 1)
```

```python
df
```

```python
df['genres'] = df['genres'].replace({'Biography': 'Drama'})
df['genres'] = df['genres'].replace({'News': 'Comedy'})
df['genres'] = df['genres'].replace({'Mystery': 'Drama'})
df['genres'] = df['genres'].replace({'Action': 'Action & Adventure'})
df['genres'] = df['genres'].replace({'Adventure': 'Action & Adventure'})
df['genres'] = df['genres'].replace({'Crime': 'Suspense'})
```

```python
df.head()
```

# VISUALIZATION

`MOST LIKELY GENRES`

```python
genre_counts = df['genres'].value_counts()

# Visualización
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='Blues')
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.title('Number of Movies by Genre')
plt.xticks(rotation=45)
plt.show()
```

`MOST LIKELY DIRECTORS`

```python
df_directors = df['director'].value_counts().head(10)
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x=df_directors, y=df_directors.index, palette='Reds') # Index is for show the how many. 
plt.title('Top directors')
plt.xlabel('Directors')
plt.ylabel('Count')

# Adjust the angle of the x-axis labels
plt.xticks(rotation=90, ha='center')

plt.show()
```
